# The error increases as the step-size increases. This makes sense, as the method is less and less accurate at approximating as the step-size increases, resulting in difference values that are not actually representative of the function.

## I used Gemini to help me formulate a function for calculating the error between the first and current run for each step size. 

```python
import numpy as np
import matplotlib.pyplot as plt
# Define parameters
steps = np.arange(0.01, 1.05, 0.05)
x_start, x_end = 0, 2
h1 = steps[0] # 0.01 
firsty = 0

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y

for idx, h in enumerate(steps) :
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    # Initial condition
    y_values[0] = 1 # y(0) = 1
    y_exact = np.exp(-2 * x_values)
    
    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
        
    if idx == 0:
        firsty = y_values

    global_error = abs(y_values[-1] - y_exact)
    step_ratio = int(round(h / h1))
    y_matches = firsty[::step_ratio]
    min_len = min(len(y_values), len(y_matches))
    point_errors = np.abs(y_values[:min_len] - y_matches[:min_len])

    # --- PRINTING RESULTS ---
    print(f"step size: h = {h:.2f}")
    print(f"total accumulated error at x={x_end}: {global_error[-1]:.6f}")
    print("current vs first run errors:")
    
    for i in range(min_len):
        print(f"  x = {x_values[i]:.2f} | Error = {point_errors[i]:.6f}")
    
    # Plot the results
    plt.plot(x_values, y_values, label="Euler's Method Approximation")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Approximation of dy/dx = -2y using Euler's Method")
    plt.legend()
    plt.show()
```
