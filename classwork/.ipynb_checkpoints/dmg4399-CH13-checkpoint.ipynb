# Matrix Multiplication Code

```python
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
```

```python
# Defining the Matrices and looking at time complexities
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt
size = 3
naiveTime = []
numpyTime = []
sizeArrays = []
for i in range(1, 9):
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)
    # find the time it takes for the naive function to run 
    naiveStart = time.perf_counter()
    placeholder = naive_matrix_multiplication(A, B)
    naiveEnd = time.perf_counter()
    naiveTime.append(naiveEnd - naiveStart)
    #find the time it takes for the numpy function to run (np.dot)
    npStart = time.perf_counter()
    placeholder = np.dot(A, B)
    npEnd = time.perf_counter()
    numpyTime.append(npEnd - npStart)
    sizeArrays.append(size)
    size *= 2
```

```python
plt.plot(sizeArrays, naiveTime, linestyle='-', marker='o', label="Naive")
plt.plot(sizeArrays, numpyTime, linestyle='-', marker='o', label="Numpy")
plt.title("Comparison of Naive and Numpy Runtime for Matrix Multiplication")
plt.legend()
plt.xlabel("Array size")
plt.ylabel("time")
plt.yscale('log')
plt.show()
```

## Comparing the two, the naive function appears to increase in the amount of time it takes to run rapidly, reaching 10s to 100s of seconds while numpy increases much slower, only growing to about 0.01 seconds in the amount of time it takes to run the multiplication. This is perfectly consistent with the time complexities, as the naive function has a complixity of O(n^3), while the numpy function generally has a time complexity lower than O(n^3) as it is highly optimized.
