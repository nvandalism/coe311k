```python
def checkDiagDom(matrix):
    for i in range(len(matrix)):
        sumofrow = 0
        for j in range(len(matrix[0])):
            if(i != j):
                sumofrow += abs(matrix[i][j])
        if (sumofrow >= matrix[i][i]):
            return False
    return True
            
```

```python
testMatrix = ([[3, 1], [1, 3]])
testMatrix2 = ([[2, 1], [1, 1]])

print(testMatrix2)
print("Is the matrix diagonally dominant?" + ' ' + str(checkDiagDom(testMatrix2)))
print(testMatrix)
print("Is the matrix diagonally dominant?" + ' ' + str(checkDiagDom(testMatrix)))
```
