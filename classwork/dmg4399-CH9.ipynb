# least squares method

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

A = np.vstack([X, np.ones(len(X))]).T

```

```python
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
plt.scatter(X, Y)
Yb = .95*X + .59
plt.plot(X, Yb)
#omg a fit line
```

# random vector soln

```python
import random
x = np.random.rand(50)
y = np.random.rand(50)
A = np.vstack([x, np.ones(len(x))]).T
```

```python
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
plt.scatter(x, y)
yb = -0.1414685872789948*x + 0.5125624948039319
plt.plot(x, yb)
#omg a fit line
```

```python
x = np.arange(0, 1.01, 0.01)
m = 2.532
b= -0.263
y = m*x + b
noise = np.random.uniform(-0.3, 0.3, len(x))
y += noise
```

```python
A = np.vstack([x, np.ones(len(x))]).T
mfit, cfit = np.linalg.lstsq(A, y, rcond=None)[0]
print(f"Slope: {mfit}, Intercept: {cfit}")
```

```python
plt.scatter(x, y)
yb = mfit*x + cfit
plt.plot(x, yb)
#omg a fit line
```
