```python
def findPolynomialAndResiduals(degree):
    coefficients = np.polyfit(x, y, degree)
    polynomial = np.poly1d(coefficients)
    y_pred = polynomial(x)
    mse = np.mean((y - y_pred)**2)
    ss_total = np.sum((y - np.mean(y))**2)
    ss_res = np.sum((y - y_pred)**2)
    r_squared = 1 - (ss_res / ss_total)
    return polynomial, mse, r_squared, y_pred
```

```python
degrees = np.linspace(1, 30, 30)
for i in degrees:
    poly, mse, r2, y_pred = findPolynomialAndResiduals(i)
    print(f"Degree {i} | MSE: {mse:.4f} | R-Squared: {r2:.4f}")
    plt.figure(figsize=(4, 3)) # Makes the plot small
    plt.scatter(x, y, s=10, color='blue', alpha=0.5, label='Data')
    plt.plot(x, y_pred, color='red', label='Fit')
    plt.legend(prop={'size': 8})
    plt.show()
```

```python
import numpy as np
# Example data points
np.random.seed(20)
x = np.random.rand(50)
y = np.random.rand(50)
# Create Vandermonde matrix for a polynomial of degree 2
degree = 3
V = np.vander(x, N=degree+1, increasing=True)
# V.T * V
VT_V = np.dot(V.T, V)
# V.T * y
VT_y = np.dot(V.T, y)
# Solve for a (coefficients)
coefficients = np.linalg.solve(VT_V, VT_y)
polynomial = np.poly1d(coefficients[::-1])
print(polynomial)
```

```python
y_pred = polynomial(x)
plt.scatter(x, y, color='blue', label='data', alpha=0.6)
sorted_X = np.argsort(x)
plt.plot(x[sorted_X], y_pred[sorted_X], color='red', linewidth=2, label='fitted polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vandermonde fit')
plt.legend()
plt.show()
```

```python
mse = np.mean((y - y_pred)**2)
print(f'Mean Squared Error: {mse:.4f}')
# R-squared value for cubic
ss_total = np.sum((y - np.mean(y))**2)
ss_res = np.sum((y - y_pred)**2)
r_squared = 1 - (ss_res / ss_total)
print(f'R-squared: {r_squared:.4f}')
```
