# Jacobi 

```python
import numpy as np
import time
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
# Example usage
A = np.array([[10, -2, 1, 0, 0],
    [-1, 11, -1, 3, 0],
    [2, -1, 10, -1, 0],
    [0, 3, -1, 8, -2],
    [0, 0, 0, -2, 9]])
b = np.array([14, 30, 26, 25, 37])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
start = time.perf_counter()
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
end = time.perf_counter()
total = end - start
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Operation Time: {total:.6e} seconds")
```

# Gauss-Seidel

```python
import numpy as np
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
# Example usage
A = np.array([[10, -2, 1, 0, 0],
    [-1, 11, -1, 3, 0],
    [2, -1, 10, -1, 0],
    [0, 3, -1, 8, -2],
    [0, 0, 0, -2, 9]])
b = np.array([14, 30, 26, 25, 37])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
start = time.perf_counter()
solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)
end = time.perf_counter()
total = end - start
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Operation Time: {total:.6e} seconds")
```

## The Gauss-Seidel method converges faster than the Jacobi method when attempting to reach a solution. Additionally, Gauss-Seidel required fewer passes and vectors decayed faster than when compared to Jacobi.
