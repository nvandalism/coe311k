```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
```

```python
print(v1, v2)
```

```python
print(v1 + v2)
print(v1 - v2)
```

```python
v_dot = np.dot(v1, v2)
print(v_dot)
```

```python
print(np.linalg.norm(v1))
```

```python
susamogus = np.array([[1, 2, 3], [4, 5, 6]])
TheOnePissIsReal = np.array([[3, 4, 3],[3,5,7],[-1, -2, -6]])
ECRA = np.array([[27, 0, 0], [4, 0, 0]])
```

```python
print(susamogus + ECRA)
print(ECRA - susamogus)
```

```python
print(np.dot(susamogus, TheOnePissIsReal))
```

```python
print(susamogus.T)
```

```python
A = np.array([[2, 3], [4, -1]])
B = np.array([5, 3])

print(np.linalg.solve(A, B))
```

```python
C = np.array([[2, 1], [1, 2]])
eigen, eigvec = np.linalg.eig(C)
print(eigen, eigvec)
```

```python
D = np.array([[4, 7], [2, 5]])
Dinv = np.linalg.inv(D)
identity = np.dot(D, Dinv)
print(identity)
```
