```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 2 * np.pi
N = 30  
x = np.linspace(a, b, N)
h = (b - a) / (N - 1)
derivative = np.cos(x)
```

```python
central = (np.sin(x + h) - np.sin(x - h)) / (2 * h)
forward = (np.sin(x + h) - np.sin(x)) / h
backward = (np.sin(x) - np.sin(x - h)) / h
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

axs[1, 0].plot(x, derivative, label='cos(x)', color='black', linewidth=2)
axs[1, 0].plot(x, central, label='ctrl diff', linestyle='--', color='green')
axs[1, 0].set_title('Central Difference')
axs[1, 0].legend()
axs[1, 0].grid(True)

axs[0, 0].plot(x, derivative, label='cos(x)', color='black', linewidth=2)
axs[0, 0].plot(x, forward, label='fwd diff', linestyle='-', color='red')
axs[0, 0].set_title('Forward Difference')
axs[0, 0].legend()
axs[0, 0].grid(True)

axs[0, 1].plot(x, derivative, label='cos(x)', color='black', linewidth=2)
axs[0, 1].plot(x, backward, label='bwd diff', linestyle='--', color='blue')
axs[0, 1].set_title('Backward Difference')
axs[0, 1].legend()
axs[0, 1].grid(True)

axs[1, 1].plot(x, derivative, label='cos(x)', color='black', linewidth=2)
axs[1, 1].plot(x, forward, label='fwd diff', linestyle=':', color='red')
axs[1, 1].plot(x, backward, label='bwd diff', linestyle='-.', color='blue')
axs[1, 1].plot(x, central, label='ctrl diff', linestyle='--', color='green')
axs[1, 1].set_title('Comparison of All Methods')
axs[1, 1].legend()
axs[1, 1].grid(True)

plt.show()
```
