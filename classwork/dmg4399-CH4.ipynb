```python
import numpy as np

def leibnezPi(numIterations):
    summation = 0
    for i in range(numIterations):
        numerator = pow(-1, i)
        denominator = 2*i + 1
        summation += 4*numerator / denominator
    return summation
```

```python
inputarray = np.arange(10, 1001, 10)
fullList = []
for i in inputarray:
    fullList.append(leibnezPi(i))
    if i%100 == 0:
        print(abs(leibnezPi(i) - np.pi))
```
