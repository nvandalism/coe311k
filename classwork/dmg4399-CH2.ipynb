```python
def convertBase10toBase2(num):
    base10 = ''
    while num/2 != 0:
        if num % 2 == 1:
            num = (num - 1) / 2
            base10 += str(1)
        else: 
            num = num / 2
            base10 += str(0)
            
    base10 = base10[::-1]
    
    return base10

print(convertBase10toBase2(286523))
print(convertBase10toBase2(2355))
```
