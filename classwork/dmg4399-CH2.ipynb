```python
def convertBase10toBase2(num):
    #string that holds the base 2 number
    base10 = ''
    #checks if the num has been reduced to 0
    while num/2 != 0:
        #Checks if the number has a remainder. If it does, append 1. If not, append 0
        if num % 2 == 1:
            num = (num - 1) / 2
            base10 += str(1)
        else: 
            num = num / 2
            base10 += str(0)
    
    #flips the string
    base10 = base10[::-1]
    
    return base10

print(convertBase10toBase2(2025))
print(convertBase10toBase2(235539))
```
