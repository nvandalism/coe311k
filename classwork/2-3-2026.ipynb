```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4, 201)

plt.subplot(2, 2, 1) 
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'g')
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x), 'b')
plt.subplot(2, 2, 4)
plt.plot(x, x*x**2*x**3, 'y')
plt.show()
```

```python
x = np.random.uniform(-4, 4, 200)
y = np.random.uniform(-4, 4, 200)
plt.scatter(x, y)
plt.colorbar()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

z1 = np.ones([2001, 2001])
im = plt.imshow(z1, cmap='Blues')
fig = plt.colorbar(im, orientation='horizontal')
plt.show(fig)
```

```python
def f1(x,y):
    return (x*x+y*y)

x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 70, cmap='hot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(45, 45)
```
