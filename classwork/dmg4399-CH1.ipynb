```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import norm

x1 = np.linspace(1, 8, 1000)
y1 = np.e**x1
y2 = x1**4
plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'b-')
plt.plot(x1, y2, 'r--')

plt.subplot(2, 2, 2)
g = np.linspace(0, 100, 1000)
h = np.linspace(0, 100, 1000)
G, H = np.meshgrid(g, h)
z1 = G*H
im = plt.imshow(z1, cmap='hot')
fig = plt.colorbar(im, orientation='horizontal')

plt.subplot(2, 2, 3)
y2 = np.random.normal(50, 10, 1000)
plt.hist(y2, 30)

plt.subplot(2, 2, 4)
x4 = np.random.uniform(-4, 4, 200)
y4 = np.random.uniform(-4, 4, 200)
size = np.random.rand(200) * 50
color = np.random.rand(200)
plt.scatter(x4, y4, size, color)
plt.colorbar()
plt.show()
```
