# Taylor Series of sin(x)

```python
#import relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import math

#Function that approximates the taylor series sin(x)
def TaylorSeriesApprox(numTerms, X):
    #Define a sum
    summation = 0
    #the function that calculates the summation function for the taylor series of sin(x)
    for i in range(numTerms):
        numerator = (-1)**i * X**(2*i + 1)
        denominator = math.factorial(2*i + 1)
        summation += numerator / denominator
    #returns the summation
    return summation
```

## Finding approximate values of sin(x) for truncated series

```python
#Step Values Used
values = np.arange(0, 10.1, .1)
#Approximations of sin(x) from 0 to 10 in steps of 0.1 stored as vectors
Approximate_Step1 = TaylorSeriesApprox(1, values)
Approximate_Step2 = TaylorSeriesApprox(2, values)
Approximate_Step3 = TaylorSeriesApprox(3, values)
Approximate_Step4 = TaylorSeriesApprox(4, values)
#Actual value of sin(x) stored as a vector
actualValues = np.sin(values)
```

## Calculate Absolute Error

```python
#Absolute error calculations stored as vectors
Error_Step1 = abs(Approximate_Step1 - actualValues)
Error_Step2 = abs(Approximate_Step2 - actualValues)
Error_Step3 = abs(Approximate_Step3 - actualValues)
Error_Step4 = abs(Approximate_Step4 - actualValues)
```

## Plotting Error vs. function of number of terms

```python
#Plot of values vs. each truncated error
plt.plot(values, Error_Step1, label='Step 1')
plt.plot(values, Error_Step2, label='Step 2')
plt.plot(values, Error_Step3, label='Step 3')
plt.plot(values, Error_Step4, label='Step 4')
plt.legend()
plt.xlabel('steps')
plt.ylabel('Absolute Error')
plt.title('sin(x) Taylor Series Truncation Error')
plt.show()
```

# Logarithmic Graph

```python
#Plot of values vs. each truncated error using the logarithmic y-scale to better represent error data
plt.plot(values, Error_Step1, label='Step 1')
plt.plot(values, Error_Step2, label='Step 2')
plt.plot(values, Error_Step3, label='Step 3')
plt.plot(values, Error_Step4, label='Step 4')
plt.legend()
plt.yscale('log')
plt.xlabel('steps')
plt.ylabel('Absolute Error')
plt.title('sin(x) Taylor Series Truncation Error')
plt.show()
```

## When we add more terms, the truncation error usually decreases because each additional term that is inputted into the taylor series fits better and better to the sin(x) curve. However, the error does not decrease uniformly for all values of x. As x increases towards 10, the error appears to grow exponentially, even though errors closer to 0 decrease. For smaller values of X, the truncation error decreases rapidly as the series converges tightly. however, for larger values of X, the term in the numerator grows much more rapidly than the denominator, resulting in higher truncation error that grows exponentially as higher and higher values of X are introduced.
