```python
import numpy as np
import matplotlib.pyplot as plt
import time
#function for the jacobi method
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    error = []
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        #add the error of the method
        error.append(np.linalg.norm(x_new - x, ord=np.inf))
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k, error
        x = x_new
    return x, max_iterations, error
```

```python
import numpy as np
#function for the gauss seidel method
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    error = []
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        #add the error of the method
        error.append(np.linalg.norm(x_new - x, ord=np.inf))
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k, error
        x = x_new
    return x, max_iterations, error
```

```python
#function initialization
A = np.array([
    [3, -0.1, -0.2],
    [0.1, 7, -0.3],
    [0.3, -0.2, 10]
])

B = np.array([7.85, -19.3, 71.4])

x0 = np.array([0.0, 0.0, 0.0])
```

```python
#error graphs for the jacobi methods in increasing tolerance
sol3, iterations3, error3 = jacobi(A, B, x0, 1e-9, 100)
print(f"tol=1e-9: {sol3=}, {iterations3=}")
plt.plot(np.arange(0, len(error3)), error3, 'b')
sol2, iterations2, error2 = jacobi(A, B, x0, 1e-6, 100)
print(f"tol=1e-6: {sol2=}, {iterations2=}")
plt.plot(np.arange(0, len(error2)), error2, 'g')
sol1, iterations1, error1 = jacobi(A, B, x0, 1e-3, 100)
print(f"tol=1e-3: {sol1=}, {iterations1=}")
plt.plot(np.arange(0, len(error1)), error1, 'r')

plt.show()
```

```python
#error graphs for the gauss_seidel method for increasing tolerance
sol3, iterations3, error3 = gauss_seidel(A, B, x0, 1e-9, 100)
print(f"tol=1e-9: {sol3=}, {iterations3=}")
plt.plot(np.arange(0, len(error3)), error3, 'b')
sol2, iterations2, error2 = gauss_seidel(A, B, x0, 1e-6, 100)
print(f"tol=1e-6: {sol2=}, {iterations2=}")
plt.plot(np.arange(0, len(error2)), error2, 'g')
sol1, iterations1, error1 = gauss_seidel(A, B, x0, 1e-3, 100)
print(f"tol=1e-3: {sol1=}, {iterations1=}")
plt.plot(np.arange(0, len(error1)), error1, 'r')

plt.show()
```

# as the tolerance gets tighter (decrease the amount), the numerical method resolves more decimal places of accuracy, increasing the accuracy of the solution. This relationship is logarithmic, as evidenced by the graph
