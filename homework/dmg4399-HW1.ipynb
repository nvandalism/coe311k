### Monte Carlo Pi Simulation 

```python
#import all required libraries to run the function and homework
import numpy as np
import matplotlib.pyplot as plt
import random

#Pi approximate monte carlo function
def approximatePi(numShots):
    #counts the number of "hits" in the region
    counter = 0
    #generates random numbers and checks if it hits within the circle
    for i in range(0, numShots):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        if (x*x + y*y < 1):
            counter += 1
    #returns the value of approximated PI
    return 4.0 * counter / numShots
```

### Code for generating number of samples from 10 to 1000 in intervals of 20

```python
#Define the arrays for plot inputs, with x being the number of samples and y being the error bound
ErrorIterations = []
numIterations = np.arange(10, 10001, 20)
numIterations = np.append(numIterations, 10000)

#Run a for loop to determine the approximation of Pi and the error of the approximation, then append it to an array
for i in numIterations:
    approximate = approximatePi(i)
    ErrorIterations.append(abs(approximate  - np.pi))
    
```

### Error Plot

```python
#Code to plot the Absolute Error as a function of the Number of Iterations
plt.plot(numIterations, ErrorIterations)
plt.title("Absolute Error between Monte Carlo Pi Approximation and true value of Pi")
plt.xlabel("Number of Iterations")
plt.ylabel("Absolute Error")
plt.show()
```
